```python
def load_images(images_dir_path):
    images = []
    image_paths = os.listdir(images_dir_path)
    images = [np.array(Image.open(f'{images_dir_path}/{image_path}')) for image_path in image_paths]
    return images
```

### Implement the MLP model

```python
###

class MLPModel(nn.Module):
    
    def __init__(self, input_dim, hidden_dim):
        super(MLPModel, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, 2)
        )
    
    def forward(self, input):
        input = input.view(input.size(0), -1)
        return self.layers(input)
    
MPL_model_instance = MLPModel(input_dim=4096, hidden_dim=128)
```
